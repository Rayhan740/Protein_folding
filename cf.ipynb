```python
import numpy as np
import matplotlib.pyplot as plt

def nt(n):
    return 0.15*(n-1.49)**4

def ntt(n):
    return 0.3*(n-1.49)**4

num = np.array([6,8,10,12,14])
cf = nt(num)
cff = ntt(num)

plt.plot(num, cf, ":", marker='o')
plt.plot(num, cff, ":", marker='o')
plt.xlabel(r'Number of monomers N')
plt.ylabel("Number of pauli strings $N_{t}$")
plt.legend(['$a(N-b)^4$','$2a(N-b)^4$'])
```

```python
import numpy as np
import matplotlib.pyplot as plt

def nt(n):
    return 4*(n-3)

def ntt(n):
    return 2*(n-3)

num = np.array([6,8,10,12,14])
cf = nt(num)
cff = ntt(num)

plt.plot(num, cf, ":", marker='o')
plt.plot(num, cff, ":", marker='o')
plt.xlabel(r'Number of monomers N')
plt.ylabel("Number of qubits")
plt.legend(['$4(N-3)$','$2(N-3)$'])
```
